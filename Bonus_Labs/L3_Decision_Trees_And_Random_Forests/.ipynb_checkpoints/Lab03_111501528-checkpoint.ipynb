```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

# load the breast_cancer dataset
init_data = load_breast_cancer()
(X, y) = load_breast_cancer(return_X_y=True)
# X = pd.DataFrame(data=X, columns=init_data['feature_names'])
# y = pd.DataFrame(data=y, columns=['label'])

bc_feature_name = init_data.feature_names
#print(init_data)
print(bc_feature_name)
print(bc_feature_name.shape[0])

# split X into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Train a RandomForestClassifier as model 
forest = RandomForestClassifier(criterion='entropy',
                                n_estimators=200, 
                                random_state=1,
                                n_jobs=2)
forest.fit(X_train, y_train)

y_pred = forest.predict(X_test)
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
print('Accuracy per feature: %.2f' % (accuracy_score(y_test, y_pred)/X.shape[1]))
```

```python
importances = forest.feature_importances_
# get sort indices in descending order
indices = np.argsort(importances)[::-1]

print(indices)

for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 50,
                            bc_feature_name[indices[f]], 
                            importances[indices[f]]))

plt.figure()
plt.title('Feature Importances')
plt.bar(range(X_train.shape[1]),
        importances[indices],
        align='center',
        alpha=0.8)

plt.xticks(range(X_train.shape[1]), 
           bc_feature_name[indices], rotation=270)
plt.tight_layout()
plt.savefig('./output/fig-forest-feature-importances.png', dpi=300)
plt.show()
```

## 1. Use the way of random forest dataset compression method

```python
for iter in range(1, bc_feature_name.shape[0]) :
    Z_forest_compression = np.empty((X.shape[0], 0))
    for i in range(iter):
        selected_column = X[:, indices[i]].reshape(-1, 1)
        Z_forest_compression = np.hstack((Z_forest_compression, selected_column))

    Z_forest_train, Z_forest_test, y_forest_train, y_forest_test = train_test_split(Z_forest_compression, y, test_size=0.3, random_state=0)
    Z_forest_forest_compression = RandomForestClassifier(criterion='entropy',
                                                     n_estimators=200,
                                                     random_state=1,
                                                     n_jobs=2)
    Z_forest_forest_compression.fit(Z_forest_train, y_forest_train)
    y_forest_forest_compression_pred = Z_forest_forest_compression.predict(Z_forest_test)
    print('Accuracy (Z_forest_forest_compression): %.2f' % accuracy_score(y_forest_test, y_forest_forest_compression_pred))
    print('Accuracy per feature (Z_forest_forest_compression): %.2f' % (accuracy_score(y_forest_test, y_forest_forest_compression_pred)/Z_forest_compression.shape[1]))
```

We can find that using this method can achieve the goal when it contains only two features, but the accuracy will drop!

## 2. Use the way of PCA method

```python
from sklearn.preprocessing import StandardScaler

# Z-normalize data
sc = StandardScaler()
Z = sc.fit_transform(X)

# Estimate the correlation matrix (R = (transpose(R)*R/N)
R = np.dot(Z.T, Z) / X.shape[0]

# Calculate the eigen values, eigen vectors
eigen_vals, eigen_vecs = np.linalg.eigh(R)

# Make a list of (eigenvalue, eigenvector) tuples
eigen_pairs = [(np.abs(eigen_vals[i]), eigen_vecs[:,i]) for i in range(len(eigen_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eigen_pairs.sort(reverse=True)

# Form the projection matrix
for iter in range(1, bc_feature_name.shape[0]):
    W_2D = np.empty((eigen_pairs[0][1].shape[0], 0))
    for i in range(iter):
        W_2D = np.hstack((W_2D, eigen_pairs[i][1][:, np.newaxis]))
        
    Z_pca_compression = Z.dot(W_2D)
    Z_pca_train, Z_pca_test, y_pca_train, y_pca_test = train_test_split(Z_pca_compression, y, test_size=0.3, random_state=0)
    Z_forest_pca_compression = RandomForestClassifier(criterion='entropy',
                                                      n_estimators=200,
                                                      random_state=1,
                                                      n_jobs=2)

    Z_forest_pca_compression.fit(Z_pca_train, y_pca_train)
    y_forest_pca_compression_pred = Z_forest_pca_compression.predict(Z_pca_test)

    print('Accuracy (Z_forest_pca_compression): %.2f' % accuracy_score(y_pca_test, y_forest_pca_compression_pred))
    print('Accuracy per feature (Z_forest_pcat_compression): %.2f' % (accuracy_score(y_pca_test, y_forest_pca_compression_pred)/Z_pca_compression.shape[1]))
    print(Z_pca_compression.shape[1])
```

We can find that using PCA method can also achive the goal, and the accuracy is even more better when containing only two features!
