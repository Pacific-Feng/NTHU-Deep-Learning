## Preparing Data

Let's prepare the training and testing datasets:

````python
df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine//wine.data', header=None)

df.columns = ['Class label', 'Alcohol', 'Malic acid', 'Ash',
              'Alcalinity of ash', 'Magnesium', 'Total phenols',
              'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 
              'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']

display(df.head())

X = df.drop(labels='Class label', axis=1)
y = df['Class label']
#X, y = df.iloc[:, 1:].values, df.iloc[:, 0].values

# split X into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0)

print('# Training data points: %d' % X_train.shape[0])
print('# Testing data points: %d' % X_test.shape[0])
print('Class labels:', np.unique(y))
````

1. 使用`train_test_split`函數將特徵矩陣`X`和目標變數`y`劃分為訓練集（`X_train`和`y_train`）和測試集（`X_test`和`y_test`），其中`test_size=0.3`表示測試集占總數據的30%，`random_state=0`是隨機種子，確保每次運行時分割結果一致。

2. `np.unique(y)` 是一個 NumPy 函數，它用於返回一個數組或列表中獨特值的陣列。在這個代碼中，`y` 是目標變數的數組，代表葡萄酒中
   據中的"Class label"列。   

當你使用 `np.unique(y)` 時，它會掃描 `y` 中的所有值，找到並返回這些值中的獨特項目（即不重複的項目）。這對於理解數據中有多少
   不同的類別或標籤是很有用的。在機器學習的上下文中，這通常用於確定分類問題中有多少個不同的類別或類別標    籤。

例如，如果 `y` 代表了一個二元分類問題，可能有兩個唯一的值，比如 [0, 1]，代表兩個不同的類別。如果 `y` 代表了多類別分
   問題，那麼 `np.unique(y)` 將返回所有不同類別    的標籤。

在你的代碼中，最後一行的 `print('Class labels:', np.unique(y))` 是用來顯示目標變數 `y` 中不同類別的標籤。這有助
   於你了解你的數據集中有多少個不同的類別標籤。

## Training

We have already decided our model to be the decision trees, so let's proceed to the training step. The Scikit-learn package provides the off-the-shelf implementation of various machine learning models / algorithms, including the decision trees. We can simply use it to build a decision tree for our training set:

````python
# criterion : impurity function
# max_depth : maximum depth of tree
# random_state : seed of random number generator
tree = DecisionTreeClassifier(criterion='entropy', 
                              max_depth=3, 
                              random_state=0)
tree.fit(X_train, y_train)
````

這段代碼使用了 scikit-learn 中的 `DecisionTreeClassifier` 類來建立一個決策樹分類器。以下是代碼中的參數解釋：

- `criterion='entropy'`：這是用於衡量每個分割點的不純度的指標。在這裡，使用了 "entropy" 作為不純度衡量標準。Entropy（熵）是一種衡量混亂度或不確定性的度量，通常用於決策樹中，目標是最小化熵以找到最佳的分割。

- `max_depth=3`：這是決策樹的最大深度。決策樹在每個節點上做出分割，並在達到最大深度或達到某些其他停止條件時停止分割。在這裡，設置了最大深度為3，這意味著決策樹將最多分成3層。

- `random_state=0`：這是隨機數生成器的種子。種子的作用是確保每次運行代碼時，生成的隨機數和樹的分割方式都是一致的。這樣可以使結果可重現，方便調試和比較。

然後，`tree.fit(X_train, y_train)` 用於將訓練數據 `X_train` 和對應的目標變數 `y_train` 用於訓練決策樹分類器。模型將根據這些數據學習如何進行分類。學習過程將根據選定的不純度衡量標準和最大深度來生成決策樹模型。

一旦模型訓練完成，你就可以使用它來進行預測，通常是使用測試數據集 `X_test`。如果你有任何關於這個決策樹模型或其後續用途的問題，請隨時提出。

NOTE: you are not required to standardize the data features before building a decision tree (or a random forest) because the information gain of a cutting point does not change when we scale values of an attribute.

## Testing

Now we have a tree. Let's apply it to our testing set to see how it performs:

````python
y_pred = tree.predict(X_test)
print('Misclassified samples: %d' % (y_test != y_pred).sum())
print('Accuracy (tree): %.2f' % ((y_test == y_pred).sum() / y_test.shape[0]))

# a more convenient way to evaluate a trained model is to use the sklearn.metrics 
from sklearn.metrics import accuracy_score
print('Accuracy (tree, sklearn): %.2f' % accuracy_score(y_test, y_pred))
````

在Python中，`%` 是一種格式化字符串的方式，稱為字串插值或字串格式化。它用於將變數的值插入到字符串中的特定位置。

在你的代碼中，`%` 被用來將變數的值插入到字符串中的特定位置，這樣你可以在打印語句中顯示變數的值。例如：

```python
print('Misclassified samples: %d' % (y_test != y_pred).sum())
```

這行代碼中，`%d` 表示在這個位置插入一個整數，而 `(y_test != y_pred).sum()` 是一個整數表達式，它計算了一些被模型錯誤分類的樣本的數量。因此，`%d` 被替換為這個整數的值，並且整個字符串將被打印出來。

這種格式化字符串的方法在Python中很常見，它允許你以一個有序的方式將變數的值插入到字符串中，使代碼更容易閱讀和理解。在字符串中，`%` 的後面跟著一個格式字符（比如 `%d` 表示整數、`%f` 表示浮點數等），而在 `%` 的後面，用括號括起來的部分是要插入到字符串中的變數或表達式。

## Training

We can build a random forest by using the RandomForestClassifier in Scikit-learn:

````python
# criterion : impurity function
# n_estimators :  number of decision trees
# random_state : seed used by the random number generator
# n_jobs : number of cores for parallelism
forest = RandomForestClassifier(criterion='entropy',
                                n_estimators=200, 
                                random_state=1,
                                n_jobs=2)
forest.fit(X_train, y_train)

y_pred = forest.predict(X_test)
print('Accuracy (forest): %.2f' % accuracy_score(y_test, y_pred))
````

We get a slightly improved accuracy 98%!

NOTE: in most implementations, including the RandomForestClassifier implementation in Scikit-learn, the bootstrap sample size $M$ is equal to the number of samples $N$ in the original training set by default. For the number of features $K$ to select at each split, the default that is used in Scikit-learn (and many other implementations) is $K$=$\sqrt{D}$, where $D$ is the number of features of data points.

## Computing Feature Importance

In addition to classification, a random forest can be used to calculate the feature importance. Using a random forest, we can measure feature importance as the averaged information gain (impurity decrease) computed from all decision trees in the forest.

````python
importances = forest.feature_importances_
# get sort indices in descending order
indices = np.argsort(importances)[::-1]

for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, 
                            X.columns.values[indices[f]], 
                            importances[indices[f]]))

plt.figure()
plt.title('Feature Importances')
plt.bar(range(X_train.shape[1]),
        importances[indices],
        align='center',
        alpha=0.5)

plt.xticks(range(X_train.shape[1]), 
           X.columns.values[indices], rotation=90)
plt.tight_layout()
plt.savefig('./output/fig-forest-feature-importances.png', dpi=300)
plt.show()
````

這段代碼是用於獲取和可視化隨機森林模型中特徵的重要性（importance）。這可以幫助你了解模型認為哪些特徵對預測結果最重要。

讓我們逐步解釋這段代碼：

1. `importances = forest.feature_importances_`：這一行代碼獲取了訓練好的隨機森林模型 `forest` 中每個特徵的重要性分數。這些分數衡量了每個特徵對模型預測的貢獻程度。

2. `indices = np.argsort(importances)[::-1]`：這一行代碼使用 `np.argsort` 函數來獲取特徵重要性的排序索引。`[::-1]` 部分用於反轉排序，以使特徵按重要性遞減的順序排列。

3. 進行一個循環，打印每個特徵的名稱和對應的重要性分數。這些信息將顯示在控制台中。

4. 接下來，代碼創建一個柱狀圖來可視化特徵的重要性。它使用 Matplotlib 库繪製柱狀圖，其中 X 軸表示特徵，Y 軸表示重要性分數。柱狀圖按照特徵重要性的降序排列，所以最重要的特徵在最左邊。

5. `plt.xticks` 和 `plt.tight_layout` 用於設置 X 軸刻度標籤的顯示和布局。

6. 最後，代碼使用 `plt.savefig` 保存生成的柱狀圖為一個圖片文件。圖片將被保存到 './output/fig-forest-feature-importances.png'。

7. `plt.show()` 用於顯示生成的柱狀圖。

這樣，你可以通過這個柱狀圖來了解哪些特徵對於你的隨機森林模型的預測最為重要。通常情況下，重要性較高的特徵對模型的預測有更大的影響，因此這可以幫助你進一步理解模型的工作方式並進行特徵選擇。

From the above figure, we can see that "Flavanoids", "OD280/OD315 of diluted wines", "Proline", and "Color intensity" are the most important features to classify the Wine dataset. This may change if we choose a different number of trees $T$ in a random foreest. For example, if we set $T=10000$, then the most important feature becomes "Color intensity."

## Feature Selection

By discarding the unimportant features, we can reduce the dimension of data points and compress data. For example, $Z_{Forest}$ is a compressed 2-D dataset that contains only the most important two features "Flavanoids" and "OD280/OD315 of diluted wines:"

````python
# 只用最重要的兩個feature來作為Z_forest的data
Z_forest = X[['Flavanoids', 'OD280/OD315 of diluted wines']].values

colors = ['r', 'b', 'g']
markers = ['s', 'x', 'o']
for l, c, m in zip(np.unique(y.values), colors, markers):
    plt.scatter(Z_forest[y.values==l, 0], 
                Z_forest[y.values==l, 1], 
                c=c, label=l, marker=m)

plt.title('Z_forest')
plt.xlabel('Flavanoids')
plt.ylabel('OD280/OD315 of diluted wines')
plt.legend(loc='lower right')
plt.tight_layout()
plt.savefig('./output/fig-forest-z.png', dpi=300)
plt.show()
````

It is worth mentioning that Scikit-learn also implements a class called SelectFromModel that helps you select features based on a user-specified threshold, which is useful if we want to use the RandomForestClassifier as a feature selector. For example, we could set the threshold to 0.16 to get $Z_{Forest}$:

````python
# put prefitted RandomForestClassifier "forest" into SelectFromModel
sfm = SelectFromModel(forest, threshold=0.16, prefit=True)
Z_forest_alt = sfm.transform(X.values)

for f in range(Z_forest_alt.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, 
                            X.columns.values[indices[f]], 
                            importances[indices[f]]))
````

這段代碼使用了 scikit-learn 中的 `SelectFromModel` 類，將預先訓練好的隨機森林分類器 "forest" 用於特徵選擇。以下是代碼的詳細解釋：

1. `sfm = SelectFromModel(forest, threshold=0.16, prefit=True)`：這一行代碼創建了一個 `SelectFromModel` 的實例 `sfm`。它的目的是通過選擇特徵以優化模型性能。`forest` 是你之前訓練的隨機森林模型，`threshold=0.16` 是設置的特徵選擇閾值。如果特徵的重要性（由隨機森林模型計算）大於等於閾值，則該特徵將被保留。

2. `Z_forest_alt = sfm.transform(X.values)`：這一行代碼使用 `sfm` 對原始特徵矩陣 `X` 執行特徵選擇，並返回選定的特徵集合，存儲在 `Z_forest_alt` 中。

3. 接下來的循環遍歷選定的特徵並打印它們的排序和重要性分數。這個循環使用之前計算的 `indices` 來確保特徵以遞減的重要性順序進行打印。

這種特徵選擇方法可以幫助你識別對預測結果最具影響力的特徵。在這個例子中，使用了隨機森林模型的特徵重要性來進行選擇，並根據閾值 `threshold` 決定哪些特徵應該被保留。這有助於簡化模型，提高訓練和預測的效率，同時保留了最具信息性的特徵。

## Dimension Reduction: PCA vs. Random Forest

So far, we have seen two dimension reduction techniques: PCA and feature selection based on Random Forest. PCA is an unsupervised dimension reduction technique since it does not require the class labels; while the latter is a supervised dimension reduction technique as the labels are used for computing the information gain for each node split. However, PCA is a feature extraction technique (as opposed to feature selection) in the sense that a reduced feature may not be identical to any of the original features. Next, let's build classifiers for the two compressed datasets $Z_{PCA}$ and $Z_{Forest}$ and compare their performance:

````python
# train a decision tree based on Z_forest

Z_forest_train, Z_forest_test, y_forest_train, y_forest_test = train_test_split(
    Z_forest, y, test_size=0.3, random_state=0)

tree_forest = DecisionTreeClassifier(criterion='entropy', 
                                     max_depth=3, 
                                     random_state=0)
tree_forest.fit(Z_forest_train, y_forest_train)

y_forest_pred = tree_forest.predict(Z_forest_test)
print('Accuracy (tree_forest): %.2f' % accuracy_score(y_forest_test, y_forest_pred))
````

````python
# train a decision tree based on Z_pca

# load Z_pca that we have created in our last lab
Z_pca= np.load('./output/Z_pca.npy')
# random_state should be the same as that used to split the Z_forest
Z_pca_train, Z_pca_test, y_pca_train, y_pca_test = train_test_split(
    Z_pca, y, test_size=0.3, random_state=0)

tree_pca = DecisionTreeClassifier(criterion='entropy', 
                                  max_depth=3, 
                                  random_state=0)
tree_pca.fit(Z_pca_train, y_pca_train)

y_pca_pred = tree_pca.predict(Z_pca_test)
print('Accuracy (tree_pca): %.2f' % accuracy_score(y_pca_test, y_pca_pred))
````

As we can see, the tree grown from <font color="red">PCA-compressed data yields the same accuracy 96% as that of the tree for uncompressed data</font>. Furthermore, it performs much better than the tree grown from the selected features advised by a random forest. This shows that PCA, a feature extraction technique, is effective in preserving "information" in a dataset when the compressed dimension is very low (2 in this case). The same holds for the Random Forest classifiers:

````python
# train a random forest based on Z_forest

forest_forest = RandomForestClassifier(criterion='entropy',
                                n_estimators=200, 
                                random_state=1,
                                n_jobs=2)
forest_forest.fit(Z_forest_train, y_forest_train)

y_forest_pred = forest_forest.predict(Z_forest_test)
print('Accuracy (forest_forest): %.2f' % accuracy_score(y_forest_test, y_forest_pred))
````

````python
# train a random forest based on Z_pca

forest_pca = RandomForestClassifier(criterion='entropy',
                                n_estimators=200, 
                                random_state=1,
                                n_jobs=2)
forest_pca.fit(Z_pca_train, y_pca_train)

y_pca_pred = forest_pca.predict(Z_pca_test)
print('Accuracy (forest_pca): %.2f' % accuracy_score(y_pca_test, y_pca_pred))
````

## Further Visualization

When the data dimension is 2, we can easily plot the decision boundaries of a classifier. Let's take a look at the decision boundaries of the Decision Tree and Random Forest classifiers we have for $Z_{PCA}$ and $Z_{Forest}$. First, we define a utility function for plotting decision boundaries:

````python
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[: len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))

    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(
            x = X[y == cl, 0],
            y = X[y == cl, 1],
            alpha = 0.8,
            c = [cmap(idx)], # Prevents warning
            marker = markers[idx],
            label = cl
        )

    # highlight test samples
    if test_idx:
        # plot all samples
        X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c=[cmap(i - 1) for i in y_test],
                    alpha=1.0,
                    linewidths=1,
                    marker='o',
                    s=55, label='test set', edgecolors='k')
````

````python
# plot boundaries of tree_forest
Z_forest_combined = np.vstack((Z_forest_train, Z_forest_test))
y_forest_combined = np.hstack((y_forest_train, y_forest_test))
plot_decision_regions(
        Z_forest_combined, 
        y_forest_combined, 
        classifier = tree_forest, 
        test_idx = range(
            y_forest_train.shape[0],
            y_forest_train.shape[0] + y_forest_test.shape[0]
        )
)

plt.title('Tree_forest')
plt.xlabel('Flavanoids')
plt.ylabel('OD280/OD315 of diluted wines')
plt.legend(loc='lower right')
plt.tight_layout()
plt.savefig('./output/fig-boundary-tree-forest.png', dpi=300)
plt.show()

# plot boundaries of tree_pca
Z_pca_combined = np.vstack((Z_pca_train, Z_pca_test))
y_pca_combined = np.hstack((y_pca_train, y_pca_test))
plot_decision_regions(
    Z_pca_combined, 
    y_pca_combined, 
    classifier = tree_pca, 
    test_idx = range(
        y_pca_train.shape[0],
        y_pca_train.shape[0] + y_pca_test.shape[0]
    )
)

plt.title('Tree_pca')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.legend(loc='lower right')
plt.tight_layout()
plt.savefig('./output/fig-boundary-tree-pca.png', dpi = 300)
plt.show()
````

As we can see, the decision boundaries of a decision tree are always axis-aligned. This means that if a "true" boundary is not axis-aligned, the tree needs to be very deep to approximate the boundary using the "staircase" one. We can see this from the random forests:

````python
# plot boundaries of tree_forest
plot_decision_regions(Z_forest_combined, 
                      y_forest_combined, 
                      classifier=forest_forest, 
                      test_idx=range(y_forest_train.shape[0],
                                     y_forest_train.shape[0] + y_forest_test.shape[0]))

plt.title('Forest_forest')
plt.xlabel('Flavanoids')
plt.ylabel('OD280/OD315 of diluted wines')
plt.legend(loc='lower right')
plt.tight_layout()
plt.savefig('./output/fig-boundary-forest-forest.png', dpi=300)
plt.show()

# plot boundaries of tree_pca
plot_decision_regions(Z_pca_combined, 
                      y_pca_combined, 
                      classifier=forest_pca, 
                      test_idx=range(y_pca_train.shape[0],
                                     y_pca_train.shape[0] + y_pca_test.shape[0]))

plt.title('Forest_pca')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.legend(loc='lower right')
plt.tight_layout()
plt.savefig('./output/fig-boundary-forest-pca.png', dpi=300)
plt.show()
````

## 2. Use the way of PCA method

````python
from sklearn.preprocessing import StandardScaler

# Z-normalize data
sc = StandardScaler()
Z = sc.fit_transform(X)

# Estimate the correlation matrix (R = (transpose(R)*R/N)
R = np.dot(Z.T, Z) / X.shape[0]

# Calculate the eigen values, eigen vectors
eigen_vals, eigen_vecs = np.linalg.eigh(R)

# Make a list of (eigenvalue, eigenvector) tuples
eigen_pairs = [(np.abs(eigen_vals[i]), eigen_vecs[:,i]) for i in range(len(eigen_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eigen_pairs.sort(reverse=True)

# Form the projection matrix
W_2D = np.hstack((eigen_pairs[0][1][:, np.newaxis],
                  eigen_pairs[1][1][:, np.newaxis]))

#print(W_2D)

Z_pca_compression = Z.dot(W_2D)
print(Z_pca_compression)

Z_pca_train, Z_pca_test, y_pca_train, y_pca_test = train_test_split(Z_pca_compression, y, test_size=0.3, random_state=0)

Z_forest_pca_compression = RandomForestClassifier(criterion='entropy',
                                                  n_estimators=200,
                                                  random_state=1,
                                                  n_jobs=2)

Z_forest_pca_compression.fit(Z_pca_train, y_pca_train)
y_forest_pca_compression_pred = Z_forest_pca_compression.predict(Z_pca_test)

print('Accuracy (Z_forest_pca_compression): %.2f' % accuracy_score(y_pca_test, y_forest_pca_compression_pred))
print('Accuracy per feature (Z_forest_pcat_compression): %.2f' % (accuracy_score(y_pca_test, y_forest_pca_compression_pred)/Z_pca_compression.shape[1]))
````

We can find that using PCA method can actually increase the accuracy and achive the goal!
